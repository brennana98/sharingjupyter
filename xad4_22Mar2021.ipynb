```python
Ntot = np.zeros(2048)
listFiles = ["xad4_1.dat.txt","xad4_2.dat.txt","xad4_3.dat.txt","xad4_4.dat.txt","xad4_5.dat.txt","xad4_6.dat.txt","xad4_7.dat.txt","xad4_8.dat.txt","xad4_9.dat.txt","xad4_10.dat.txt","xad4_11.dat.txt","xad4_12.dat.txt","xad4_13.dat.txt","xad4_14.dat.txt","xad4_15.dat.txt"]

for i in listFiles:
    data = open(i,"r")
    ch,N = np.loadtxt(data,unpack=True)
    data.close
    Ntot = Ntot + N
    
#print(Ntot)

npts = len(Ntot)
ngood = np.zeros(2048)
ngood[0:npts-1] = Ntot[npts-1:0:-1]
#print(ngood)

#print(len(ch))
#print(len(ngood))

#for loop including each value from the different 16 runs, adding number of counts, N, to Ntot within loop
#lower channels correspond to longer times, data is backwards
#switch indices using length (npts = len(Ntot) and ngood[0:npts-1] = Ntot[npts-1:0:-1])
#read files () == ch, N
#start point, mark as time zero, keep time as an offset, take away some reference point of start of data

#Plot data
plt.figure
plt.xlabel("ch")
plt.ylabel(r"$N$")
plt.xlim([200,1000])
plt.yscale('log')
plt.ylim([10,600000])
plt.plot(ch,ngood)
#plt.errorbar(ch,ngood,xerr=None,yerr=dnpw,fmt='b.',capsize=1)
plt.show()

# conversion for channels/ns
tc = 1.96

#channel for t=0
b = 470. 

#constant background - fixed
#d=np.mean(npw[400:450])

#fitting function: 3 exponentials + constant background
def fitfn(xin,a1,c1,a2,c2,a3,c3,d):
    return a1*np.exp((xin-b)/(c1*tc))+a2*np.exp((xin-b)/(c2*tc))+a3*np.exp((xin-b)/(c3*tc))+d

#initial guess at parameters
a1 = 360000.
c1 = -2.3
a2 = 10000.
c2 = -10.
a3 = 750.
c3 = -100.
d = 100.
#print(d)

#Plot initial guess at fitting function compared to data
plt.figure
plt.xlabel("ch")
plt.ylabel(r"$N$")
plt.xlim([400,2000])
plt.yscale('log')
plt.ylim([10,200000])
plt.plot(ch,ngood)
plt.plot(ch,fitfn(ch,a1,c1,a2,c2,a3,c3,d))
plt.show()
```

```python
#choose region of data to fit
peakcut=np.logical_and(ch>500,ch<1801)
peakind=np.where(peakcut)
chf = ch[peakind]
nptsf=len(chf)
nf = ngood[peakind]
dnf = np.sqrt(ngood)[peakind]

#define parameter array and do fit
p0=np.array([a1,c1,a2,c2,a3,c3,d])
pfit,pcov=scip.optimize.curve_fit(fitfn,chf,nf,p0,dnf,absolute_sigma=True)#####this is important and needed!!
perr=np.sqrt(np.diag(pcov))
np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
print(pfit)
#print("%8.2f" % (pfit[1]))
print(perr)

#update parameters with best fit values
a1f=pfit[0]
c1f=pfit[1]
a2f=pfit[2]
c2f=pfit[3]
a3f=pfit[4]
c3f=pfit[5]
df=pfit[6]


#calculate residuals, chisquared
resid=nf-fitfn(chf,a1f,c1f,a2f,c2f,a3f,c3f,df)
chisq=np.sum((resid/dnf)**2)
dof=nptsf-len(p0)
chisqr=chisq/dof
CDF=chi2.cdf(chisq,dof)
print("xad4: chisq={:8.2f} DOF={:5d} chisqr={:6.2f} CL={:10.7f}".format(chisq,dof,chisqr,100.*(1.-CDF)))

#For plotting, parameters of best-fit function
pfitpw=np.array([a1f,c1f,a2f,c2f,a3f,c3f,df,chisqr])
perrpw=perr
#xplot=np.linspace(300,1500)

#plot data and best fit function
plt.figure(1,figsize=(10,4))
plt.subplot(1,2,1)
plt.xlabel("ch")
plt.ylabel(r"$N_{xad4}$")
plt.xlim([300,2000])
plt.yscale('log')
plt.ylim([10,10000])
plt.plot(chf,nf)
plt.plot(chf,fitfn(chf,a1f,c1f,a2f,c2f,a3f,c3f,df))
#plot residuals
plt.subplot(1,2,2)
plt.xlabel("ch")
plt.xlim([500,700])
plt.yscale('linear')
plt.ylim([-200,200])
plt.errorbar(chf,resid,xerr=None,yerr=dnf,fmt='b.',capsize=2)
plt.show()
```
